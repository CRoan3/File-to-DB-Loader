```python
def get_column_names(schemas, ds_name, sorting_key = 'column_position'):
    column_details = schemas[ds_name] #ds_name = dataset name
    columns = sorted(column_details, key = lambda col: col[sorting_key]) 	#sorting columns by specified sorting_key
    return [col['column_name'] for col in columns] #gives column names
```

```python
%%sql
TRUNCATE TABLE orders, order_items
```

```python
df_reader = pd.read_csv(f'data/retail_db/orders/part-00000', names = columns, chunksize=10000) #breaks up df into chunks of 10000. we have to run this every time we want to work with it because the data will be gone every time process it
```

```python
for idx, df in enumerate(df_reader):
    print(f'Size of chunk {idx} is {df.shape}')
```

Enumerate gives index for each and every df that is part os df_reader as we iterate through it. Enumerate is a function we use at the time of looping to get the indexes so that we do not need to use counters

```python
for idx, df in enumerate(df_reader):
    print(f'Processing chunk {idx} with size {df.shape} of {ds_name}')
    df.to_sql(
        'orders', 
        conn_url, 
        if_exists='append', #append vs replace - replace: drop table before inserting new values. Append: insert new values into existing table
        index=False
)
```
